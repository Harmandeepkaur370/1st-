```python
# =========================================
#  Multi-Disease Prediction App (Colab + Gradio)
# Diabetes, Heart Disease, Parkinsons Prediction
# =========================================

!pip install -q gradio scikit-learn pandas numpy

import pickle
import numpy as np
import gradio as gr
import os

# -----------------------------------------
# Load saved models
# -----------------------------------------
MODEL_PATH = '/content/saved_models/'  # Ensure your models are in this folder

diabetes_model = pickle.load(open(os.path.join(MODEL_PATH, '/content/diabetes_model.sav'), 'rb'))
heart_disease_model = pickle.load(open(os.path.join(MODEL_PATH, '/content/heart_disease_model.sav'), 'rb'))
parkinsons_model = pickle.load(open(os.path.join(MODEL_PATH, '/content/parkinsons_model.sav'), 'rb'))

# -----------------------------------------
# Define Prediction Functions
# -----------------------------------------
def predict_diabetes(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin,
                     BMI, DiabetesPedigreeFunction, Age):
    inputs = [float(x) for x in [Pregnancies, Glucose, BloodPressure, SkinThickness,
                                 Insulin, BMI, DiabetesPedigreeFunction, Age]]
    prediction = diabetes_model.predict([inputs])
    return "The person is diabetic" if prediction[0] == 1 else "The person is not diabetic"

def predict_heart_disease(age, sex, cp, trestbps, chol, fbs, restecg, thalach,
                          exang, oldpeak, slope, ca, thal):
    inputs = [float(x) for x in [age, sex, cp, trestbps, chol, fbs, restecg,
                                 thalach, exang, oldpeak, slope, ca, thal]]
    prediction = heart_disease_model.predict([inputs])
    return "The person has heart disease" if prediction[0] == 1 else "The person does not have heart disease"

def predict_parkinsons(fo, fhi, flo, Jitter_percent, Jitter_Abs,
                       RAP, PPQ, DDP, Shimmer, Shimmer_dB, APQ3, APQ5,
                       APQ, DDA, NHR, HNR, RPDE, DFA, spread1, spread2, D2, PPE):
    inputs = [float(x) for x in [fo, fhi, flo, Jitter_percent, Jitter_Abs,
                                 RAP, PPQ, DDP, Shimmer, Shimmer_dB, APQ3, APQ5,
                                 APQ, DDA, NHR, HNR, RPDE, DFA, spread1, spread2, D2, PPE]]
    prediction = parkinsons_model.predict([inputs])
    return "The person has Parkinson's disease" if prediction[0] == 1 else "The person does not have Parkinson's disease"

# -----------------------------------------
# Create Gradio Interface
# -----------------------------------------
with gr.Blocks() as demo:

    gr.Markdown("##  Multi-Disease Prediction System")

    with gr.Tabs():
        # Diabetes Tab
        with gr.TabItem("Diabetes Prediction"):
            with gr.Row():
                Pregnancies = gr.Number(label="Number of Pregnancies")
                Glucose = gr.Number(label="Glucose Level")
                BloodPressure = gr.Number(label="Blood Pressure")
            with gr.Row():
                SkinThickness = gr.Number(label="Skin Thickness")
                Insulin = gr.Number(label="Insulin Level")
                BMI = gr.Number(label="BMI")
            with gr.Row():
                DiabetesPedigreeFunction = gr.Number(label="Diabetes Pedigree Function")
                Age = gr.Number(label="Age")
            diabetes_button = gr.Button("Diabetes Test Result")
            diabetes_output = gr.Textbox(label="Result")
            diabetes_button.click(fn=predict_diabetes,
                                  inputs=[Pregnancies, Glucose, BloodPressure, SkinThickness,
                                          Insulin, BMI, DiabetesPedigreeFunction, Age],
                                  outputs=diabetes_output)

        # Heart Disease Tab
        with gr.TabItem("Heart Disease Prediction"):
            with gr.Row():
                age = gr.Number(label="Age")
                sex = gr.Number(label="Sex")
                cp = gr.Number(label="Chest Pain types")
            with gr.Row():
                trestbps = gr.Number(label="Resting Blood Pressure")
                chol = gr.Number(label="Serum Cholestoral")
                fbs = gr.Number(label="Fasting Blood Sugar > 120 mg/dl")
            with gr.Row():
                restecg = gr.Number(label="Resting Electrocardiographic results")
                thalach = gr.Number(label="Maximum Heart Rate achieved")
                exang = gr.Number(label="Exercise Induced Angina")
            with gr.Row():
                oldpeak = gr.Number(label="ST depression induced by exercise")
                slope = gr.Number(label="Slope of the peak exercise ST segment")
                ca = gr.Number(label="Major vessels colored by flourosopy")
                thal = gr.Number(label="thal (0=normal,1=fixed defect,2=reversable defect)")
            heart_button = gr.Button("Heart Disease Test Result")
            heart_output = gr.Textbox(label="Result")
            heart_button.click(fn=predict_heart_disease,
                               inputs=[age, sex, cp, trestbps, chol, fbs, restecg, thalach,
                                       exang, oldpeak, slope, ca, thal],
                               outputs=heart_output)

        # Parkinson's Tab
        with gr.TabItem("Parkinsons Prediction"):
            features = ['fo','fhi','flo','Jitter_percent','Jitter_Abs','RAP','PPQ','DDP','Shimmer',
                        'Shimmer_dB','APQ3','APQ5','APQ','DDA','NHR','HNR','RPDE','DFA','spread1','spread2','D2','PPE']
            inputs = {}
            for f in features:
                inputs[f] = gr.Number(label=f)
            parkinsons_button = gr.Button("Parkinson's Test Result")
            parkinsons_output = gr.Textbox(label="Result")
            parkinsons_button.click(fn=predict_parkinsons,
                                    inputs=[inputs[f] for f in features],
                                    outputs=parkinsons_output)

# Launch the app
demo.launch()
```

```python
!pip install gradio==4.44.0 matplotlib scikit-learn joblib

import gradio as gr
import joblib, os, random
import matplotlib.pyplot as plt

# -----------------------------
# Safe model loader with fallback
# -----------------------------
def safe_load_model(path):
    if os.path.exists(path):
        try:
            return joblib.load(path)
        except Exception as e:
            print(f"[warn] Failed to load {path}: {e}")
    print(f"[warn] Using dummy predictor for {path}")
    return lambda X: [random.choice([0, 1])]

diabetes_model = safe_load_model("/content/saved_models/diabetes_model.sav")
heart_model = safe_load_model("/content/saved_models/heart_disease_model.sav")
parkinsons_model = safe_load_model("/content/saved_models/parkinsons_model.sav")

# -----------------------------
# User store + history
# -----------------------------
USERS = {"test": "123"}  # default user
history = {}

# -----------------------------
# Prediction Functions
# -----------------------------
def predict_diabetes(preg, glu, bp, skin, ins, bmi, dpf, age, user):
    X = [[preg, glu, bp, skin, ins, bmi, dpf, age]]
    result = diabetes_model(X)[0]
    msg = " Diabetes Detected" if result == 1 else " No Diabetes"
    history.setdefault(user, []).append(("Diabetes", msg))
    return msg

def predict_heart(age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal, user):
    X = [[age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal]]
    result = heart_model(X)[0]
    msg = " Heart Disease Detected" if result == 1 else " No Heart Disease"
    history.setdefault(user, []).append(("Heart", msg))
    return msg

def predict_parkinsons(fo, fhi, flo, jitter, shimmer, rap, ppq, d2, nhr, hnr, user):
    X = [[fo, fhi, flo, jitter, shimmer, rap, ppq, d2, nhr, hnr]]
    result = parkinsons_model(X)[0]
    msg = " Parkinsons Detected" if result == 1 else " No Parkinsons"
    history.setdefault(user, []).append(("Parkinsons", msg))
    return msg

# -----------------------------
# Login / Signup / Logout
# -----------------------------
def signup_fn(username, password):
    if username in USERS:
        return " User already exists!"
    USERS[username] = password
    return " Signup successful! Please login."

def login_fn(username, password):
    if username in USERS and USERS[username] == password:
        return (
            f" Welcome {username}!",   # auth_msg
            username,                   # state
            gr.update(visible=True),    # show dashboard
            f" Hello, **{username}**!", # greet
            gr.update(visible=False)    # hide login panel
        )
    return (
        " Invalid credentials!",
        "",
        gr.update(visible=False),
        "",
        gr.update(visible=True)
    )

def logout_fn():
    return "", "", gr.update(visible=False), "", gr.update(visible=True)

# -----------------------------
# History Display
# -----------------------------
def show_history(user):
    if not user or user not in history or not history[user]:
        return " No history yet!"
    return "\n".join([f" {d}: {res}" for d, res in history[user]])

def show_pie(user):
    if not user or user not in history or not history[user]:
        return None
    counts = {}
    for d, res in history[user]:
        counts[res] = counts.get(res, 0) + 1
    labels, sizes = counts.keys(), counts.values()
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%')
    return fig

# -----------------------------
# Gradio UI
# -----------------------------
with gr.Blocks(css="""
    body {background: #f0f4f9;}
    .gr-button {font-weight:bold;}
    h1, h2 {color:#2b4c7e;}
""") as demo:
    state = gr.State("")

    # --- LOGIN PANEL ---
    with gr.Group(visible=True) as login_panel:
        gr.Markdown("<h1 style='text-align:center;'> AI Health Predictor</h1>")
        gr.Markdown("<p style='text-align:center;'>Login or signup to access your health dashboard</p>")
        user_input = gr.Textbox(label=" Username")
        pwd_input = gr.Textbox(label=" Password", type="password")
        with gr.Row():
            login_btn = gr.Button("Login ", elem_classes="gr-button")
            signup_btn = gr.Button("Signup ", elem_classes="gr-button")
        auth_msg = gr.Markdown("")

    # --- DASHBOARD ---
    with gr.Group(visible=False) as dashboard:
        greet = gr.Markdown("")
        logout_btn = gr.Button(" Logout")

        with gr.Tab(" Diabetes"):
            gr.Markdown("### Diabetes Prediction")
            di_in = [gr.Number(label=x) for x in ["Pregnancies","Glucose","BP","SkinThickness","Insulin","BMI","DPF","Age"]]
            di_btn = gr.Button(" Predict Diabetes")
            di_out = gr.Textbox(label="Result")

        with gr.Tab(" Heart Disease"):
            gr.Markdown("### Heart Disease Prediction")
            he_in = [gr.Number(label=x) for x in ["Age","Sex","CP","Trestbps","Chol","Fbs","Restecg","Thalach","Exang","Oldpeak","Slope","Ca","Thal"]]
            he_btn = gr.Button(" Predict Heart Disease")
            he_out = gr.Textbox(label="Result")

        with gr.Tab(" Parkinsons"):
            gr.Markdown("### Parkinsons Prediction")
            pa_in = [gr.Number(label=x) for x in ["Fo","Fhi","Flo","Jitter","Shimmer","RAP","PPQ","D2","NHR","HNR"]]
            pa_btn = gr.Button(" Predict Parkinsons")
            pa_out = gr.Textbox(label="Result")

        with gr.Tab(" History & Analytics"):
            his_btn = gr.Button(" Show History")
            his_out = gr.Textbox(label="History", lines=6)
            pie_btn = gr.Button(" Show Analytics")
            pie_out = gr.Plot()

    # --- EVENTS ---
    signup_btn.click(signup_fn, [user_input, pwd_input], [auth_msg])
    login_btn.click(login_fn, [user_input, pwd_input],
                    [auth_msg, state, dashboard, greet, login_panel])
    logout_btn.click(logout_fn, None, [auth_msg, state, dashboard, greet, login_panel])

    di_btn.click(predict_diabetes, di_in+[state], di_out)
    he_btn.click(predict_heart, he_in+[state], he_out)
    pa_btn.click(predict_parkinsons, pa_in+[state], pa_out)
    his_btn.click(show_history, [state], [his_out])
    pie_btn.click(show_pie, [state], [pie_out])

demo.launch()
```
